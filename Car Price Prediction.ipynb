```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#  Step 01: Load the dataset

```python
car_data = pd.read_csv("CAR DETAILS.csv")
car_data
```

```python
# Display basic information about the dataset
```

```python
print(car_data.info())
```

```python
print(car_data.head())
```

```python
# Step 1: Data Cleaning and Overview
print("\nSummary of Categorical Variables:")
for col in ['fuel', 'seller_type', 'transmission', 'owner']:
    print(f"{col}:\n{car_data[col].value_counts()}\n")
```

```python
print("\nSummary of Numerical Variables:")
print(car_data.describe())
```

# Step 02: Visualizations 

```python
# Year Distribution
plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 1)
sns.histplot(car_data['year'], bins=30, kde=True, color='blue')
plt.title('Distribution of Year')
```

```python
# Selling Price Distribution
plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 2)
sns.histplot(car_data['selling_price'], bins=30, kde=True, color='green')
plt.title('Distribution of Selling Price')
```

```python
# KM Driven Distribution
plt.figure(figsize=(20, 5))
plt.subplot(1, 3, 3)
sns.histplot(car_data['km_driven'], bins=30, kde=True, color='purple')
plt.title('Distribution of KM Driven')
plt.tight_layout()
plt.show()
```

```python
# Step 3: Visualizations - Categorical Variables
plt.figure(figsize=(14, 10))
categorical_columns = ['fuel', 'seller_type', 'transmission', 'owner']
for i, col in enumerate(categorical_columns):
    plt.subplot(2, 2, i + 1)
    sns.countplot(data=car_data, x=col, palette='viridis', order=car_data[col].value_counts().index)
    plt.title(f"Count of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# Step 03:  Relationship Analysis

```python
# Selling Price vs. Year
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(data=car_data, x='year', y='selling_price', hue='fuel', alpha=0.7, palette='deep')
plt.title("Selling Price vs. Year")
plt.xlabel("Year")
plt.ylabel("Selling Price")
```

```python
# Selling Price vs. KM Driven
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 2)
sns.scatterplot(data=car_data, x='km_driven', y='selling_price', hue='transmission', alpha=0.7, palette='muted')
plt.title("Selling Price vs. KM Driven")
plt.xlabel("Kilometers Driven")
plt.ylabel("Selling Price")
plt.xscale('log')  # Log scale for better visualization
plt.tight_layout()
plt.show()
```

```python
# Step 5: Correlation Heatmap
plt.figure(figsize=(8, 6))
correlation_matrix = car_data[['year', 'selling_price', 'km_driven']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Variables")
plt.show()
```
